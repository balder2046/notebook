```python
temp_out = 70.0
r = 0.9
```

```python
def cooling_law(temp_start, walk_time):
    return temp_out + (temp_start - temp_out) * r ** walk_time
def temperature_mix(temp_coff,vol_coff,temp_cream,vol_cream):
    return (temp_coff * vol_coff + temp_cream * vol_cream) / (vol_coff + vol_cream)
```

```python
cooling_law(185.0, 10.0)
```

```python
times = arange(0.0,21.0,5)
tempratures = cooling_law(185.0, times)
tempratures
```

```python
temp_coffee = 185.0
vol_coffee = 8.0
vol_cream = 1.0
walk_time = 10.0
r = 0.9
def temperature_difference(temp_cream, temp_out):
    temp_start = (temp_coffee * vol_coffee + temp_cream * vol_cream) / (vol_coffee + vol_cream)
    temp_mix_at_shop = (temp_out + (temp_start - temp_out) * r ** walk_time)
    temp_start = temp_out + (temp_coffee - temp_out) * r **walk_time
    temp_mix_at_office = (temp_start * vol_coffee + temp_cream * vol_cream) / (vol_coffee + vol_cream)
    return temp_mix_at_shop - temp_mix_at_office
cream_temperatures = arange(40.0,51.0,1.0)
outside_temperatures = arange(35.0,60.0,1.)
cream_values,outside_values = meshgrid(cream_temperatures,outside_temperatures)
temperature_differences = temperature_difference(cream_values, outside_values)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
fig.set_size_inches(7,5)
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(cream_values,outside_values, temperature_differences, rstride=1, cstride=1,cmap=cm.coolwarm,linewidth=0)
xlabel("Cream Temperature")
ylabel("Outside Temperature")
title('Temperature difference')
```
